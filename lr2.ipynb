# 3. Создать произвольную верхнюю унитреугольную матрицу А 5 порядка, вектор B произвольный. Решить систему AX = B.

```python
d1, d2 = 5, 5
dm, dM = 2, 10

D = np.random.randint(dm,dM,(d1, d2))
D = np.triu(D, k=0)
for i in range(D.shape[0]):
    D[i][i] = 1

E = np.random.randint(dm,dM,(d1, 1))
n = len(E)
X = np.zeros(n)


print("A:\n", D, "\n\nB:\n", E, "\n\nX:\n", X)
```

```python
# D*X=E  => если detD != 0:  D1*D*X=D1*E
print("Определитель A =", np.linalg.det(D))

X[n-1] = E[n-1, 0]
for i in range(n-2, -1, -1):
    X[i] = E[i, 0] - np.dot(D[i, i+1:], X[i+1:])

# print("X:\n", X, "\n\nAX:\n", D @ X)
print("X:", X)

print("\nЭталон:\n", np.linalg.solve(D, E))
```
